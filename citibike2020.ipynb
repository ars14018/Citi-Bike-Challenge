```python
#Dependences 
import pandas as pd
import numpy as np 
```

```python
# Take in all of our CitiBike data and read it into pandas
jan_2020 = "./Resources/JC-202001-citibike-tripdata.csv"
feb_2020 = "./Resources/JC-202002-citibike-tripdata.csv"
mar_2020 = "./Resources/JC-202003-citibike-tripdata.csv"
apr_2020 = "./Resources/JC-202004-citibike-tripdata.csv"
may_2020 = "./Resources/JC-202005-citibike-tripdata.csv"
june_2020 = "./Resources/JC-202006-citibike-tripdata.csv"
july_2020 = "./Resources/JC-202007-citibike-tripdata 2.csv"
aug_2020 = "./Resources/JC-202008-citibike-tripdata.csv"
sept_2020 = "./Resources/JC-202009-citibike-tripdata 2.csv"
oct_2020 = "./Resources/JC-202010-citibike-tripdata.csv"
nov_2020 = "./Resources/JC-202011-citibike-tripdata.csv"

#Create DataFrames for all the data files
jan_2020_df = pd.read_csv(jan_2020)
feb_2020_df = pd.read_csv(feb_2020)
mar_2020_df = pd.read_csv(mar_2020)
apr_2020_df = pd.read_csv(apr_2020)
may_2020_df = pd.read_csv(may_2020)
june_2020_df = pd.read_csv(june_2020)
july_2020_df = pd.read_csv(july_2020)
aug_2020_df = pd.read_csv(aug_2020)
sept_2020_df = pd.read_csv(sept_2020)
oct_2020_df = pd.read_csv(oct_2020)
nov_2020_df = pd.read_csv(nov_2020)
```

```python
#Merge all data from 2019 into one big data frame
combined2020_CitiBike_df = pd.merge(jan_2020_df, feb_2020_df, how='outer')
combined2020_CitiBike_df3 = pd.merge(combined2020_CitiBike_df,mar_2020_df, how = "outer")
combined2020_CitiBike_df4 = pd.merge(combined2020_CitiBike_df3,apr_2020_df, how = "outer")
combined2020_CitiBike_df5 = pd.merge(combined2020_CitiBike_df4,may_2020_df, how = "outer")
combined2020_CitiBike_df6 = pd.merge(combined2020_CitiBike_df5,june_2020_df, how = "outer")
combined2020_CitiBike_df7 = pd.merge(combined2020_CitiBike_df6,july_2020_df, how = "outer")
combined2020_CitiBike_df8 = pd.merge(combined2020_CitiBike_df7,aug_2020_df, how = "outer")
combined2020_CitiBike_df9 = pd.merge(combined2020_CitiBike_df8,sept_2020_df, how = "outer")
combined2020_CitiBike_df10 = pd.merge(combined2020_CitiBike_df9,oct_2020_df, how = "outer")
combined2020_CitiBike_df11 = pd.merge(combined2020_CitiBike_df10,nov_2020_df, how = "outer")

combined2020_CitiBike_df11
```

```python
#splitting start time column 
combined2020_CitiBike_df11['starttime'] = pd.to_datetime(combined2020_CitiBike_df11['starttime'])
combined2020_CitiBike_df11['start date'] = combined2020_CitiBike_df11['starttime'].dt.date
combined2020_CitiBike_df11['start time'] = combined2020_CitiBike_df11['starttime'].dt.time
combined2020_CitiBike_df11.head()
```

```python
#splitting end time time column 
combined2020_CitiBike_df11['stoptime'] = pd.to_datetime(combined2020_CitiBike_df11['stoptime'])
combined2020_CitiBike_df11['end date'] = combined2020_CitiBike_df11['stoptime'].dt.date
combined2020_CitiBike_df11['end time'] = combined2020_CitiBike_df11['stoptime'].dt.time
combined2020_CitiBike_df11.head()
```

```python
# Drop starttime column

combined2020_CitiBike_df11.drop('starttime', inplace=True, axis=1)
combined2020_CitiBike_df11.head()
```
